# Perceptronの実装

## 最初のAND

```python
import numpy as np
```

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1

print("AND(0,0):{}".format(AND(0,0)))
print("AND(1,0):{}".format(AND(1,0)))
print("AND(0,1):{}".format(AND(0,1)))
print("AND(1,1):{}".format(AND(1,1)))
```

## ANDの重みとバイアスによるバリエーション

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print("AND(0,0):{}".format(AND(0,0)))
print("AND(1,0):{}".format(AND(1,0)))
print("AND(0,1):{}".format(AND(0,1)))
print("AND(1,1):{}".format(AND(1,1)))
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print("NAND(0,0):{}".format(NAND(0,0)))
print("NAND(1,0):{}".format(NAND(1,0)))
print("NAND(0,1):{}".format(NAND(0,1)))
print("NAND(1,1):{}".format(NAND(1,1)))
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print("OR(0,0):{}".format(OR(0,0)))
print("OR(1,0):{}".format(OR(1,0)))
print("OR(0,1):{}".format(OR(0,1)))
print("OR(1,1):{}".format(OR(1,1)))
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print("XOR(0,0):{}".format(XOR(0,0)))
print("XOR(1,0):{}".format(XOR(1,0)))
print("XOR(0,1):{}".format(XOR(0,1)))
print("XOR(1,1):{}".format(XOR(1,1)))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

def func(X,Y):
    return -0.5*X-0.5*Y+0.7

x = np.arange(-0.5,1.5,0.25)
y = np.arange(-0.5,1.5,0.25)

X, Y = np.meshgrid(x,y)

Z = func(X,Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X,Y,Z)

ax.plot_wireframe(X,Y,0)

plt.show()
```
